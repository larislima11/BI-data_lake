```python

import pandas as pd
import numpy as np
import os

if not os.path.exists('data_lake'):
    os.makedirs('data_lake')

num_files = 10
num_rows_per_file = 1000

dfs = []

nomes = ['Carlos', 'Ana', 'João', 'Maria', 'Pedro', 'Lucas', 'Mariana', 'Beatriz', 'Felipe', 'Sofia']
cidades = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'Brasília']

for i in range(num_files):

    data = {
        'ClienteID': np.arange(i * num_rows_per_file, (i + 1) * num_rows_per_file),
        'Nome': np.random.choice(nomes, num_rows_per_file),
        'Idade': np.random.randint(18, 70, num_rows_per_file),
        'Cidade': np.random.choice(cidades, num_rows_per_file),
        'Valor_Compra': np.round(np.random.uniform(50.0, 1000.0, num_rows_per_file), 2)
    }

    # Cria uma tabela, sendo DataFrame, com os dados gerados:
    df = pd.DataFrame(data)

    # Define nomes dos arquivos para CSV, JSON e XLSX
    file_name_csv = f'data_lake/clientes_dados{i+1}.csv'
    file_name_json = f'data_lake/clientes_dados{i+1}.json'
    file_name_xlsx = f'data_lake/clientes_dados{i+1}.xlsx'

    # Salva os arquivos em CSV, JSON e XLSX no diretório 'data_lake'
    df.to_csv(file_name_csv, index=False)
    df.to_json(file_name_json, orient='records', lines=True)
    df.to_excel(file_name_xlsx, index=False)

    # Adiciona o arquivo e o DataFrame ao array:
    dfs.append((file_name_csv, file_name_json, file_name_xlsx, df))

print("DADOS DE CLIENTES PARA O DATALAKE FORAM GERADOS COM SUCESSO EM CSV, JSON E XLSX")

# Exibe as informações dos arquivos:
for file_name_csv, file_name_json, file_name_xlsx, df in dfs:
    print(f"\nINFORMAÇÕES DOS ARQUIVOS: {file_name_csv}, {file_name_json}, {file_name_xlsx}\n")
    print(df.head())

```
